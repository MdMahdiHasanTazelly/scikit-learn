```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
type(iris.data)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==0].count()
```

```python
# generating new column as flower name
df['flower_name'] = df.target.apply( lambda x: iris.target_names[x] )  # on target column's x index 
                                                                       #apply iris.target_names[x]
df.head()
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower_name'], axis='columns')
y = df.target
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2 )
print(x_train.shape, x_test.shape)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
